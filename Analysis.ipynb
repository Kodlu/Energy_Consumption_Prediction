```python
import pandas as pd
data = pd.read_csv("New_IOT_Data.csv", encoding= 'unicode_escape')
```

```python
# import required libraries
from sklearn.ensemble import ExtraTreesRegressor
import numpy as np

# create some random data for demonstration
X = np.array(data[['Total CPU Usage [%]' , 'CPU SOC [°C]']])
y = np.array(data['Core+SoC Power (SVI2 TFN) [W]'])

# create an instance of the Extra Trees Regressor
regressor = ExtraTreesRegressor(n_estimators=100, random_state=0)
```

```python
# fit the regressor to the data
regressor.fit(X, y)

# predict the target variable for some new input data
new_data = np.array([[0.2, 0.3], [0.4, 0.5]])
predictions = regressor.predict(new_data)

# print the predicted values
print(predictions)
```

```python
y_pred = regressor.predict(X)
```

```python
from sklearn.metrics import r2_score, mean_squared_error


# calculate the R-squared score
r2 = r2_score(y, y_pred)
print("R-squared score:", r2)

# calculate the mean squared error
mse = mean_squared_error(y, y_pred)
print("Mean squared error:", mse)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# assuming you have y_actual and y_predicted as numpy arrays
y_actual = y
y_predicted = y_pred

# calculate the R-squared score
r2 = r2_score(y_actual, y_predicted)
r2_percentage = r2 * 100
print("R-squared score:", r2_percentage, "%")

# calculate the mean squared error
mse = mean_squared_error(y_actual, y_predicted)
mse_percentage = mse * 100
print("Mean squared error:", mse_percentage, "%")
```

```python
import pickle
pickle.dump(regressor, open('model_regressor.pkl', 'wb'))
```
